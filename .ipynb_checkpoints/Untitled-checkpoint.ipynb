```python
def getData():
    columns = []
    solutions = []
    with open("mimic_synthetic_feat.name") as f:
        for line in f:
            columns.append(line[:-1]) #remove \n
    with open("mimic_synthetic_train.solution") as f:
        for line in f:
            solutions.append(line[:-1]) #remove \n
        solutions = solutions[:-1] # last line is empty
    df = pd.read_csv("mimic_synthetic_train.data",sep=" ",names=columns)
    df["DIED"] = solutions # 0 False 1 True
    return df
def plot_df_column_by(df,group_by,column):
    df.groupby(group_by)[column].plot(legend=True,title=column)
    plt.show()
```

```python
df =getData()
df.head()
```
